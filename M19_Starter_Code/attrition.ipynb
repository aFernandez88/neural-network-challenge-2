## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
y_df['Department'].value_counts()
```

```python
# Create a list of at least 10 column names to use as X data
X_columns = ['Age', 'Education', 'DistanceFromHome', 'JobSatisfaction',
             'OverTime', 'StockOptionLevel', 'WorkLifeBalance', 'NumCompaniesWorked',
             'YearsAtCompany', 'YearsSinceLastPromotion']



# Create X_df using your selected columns
X_df = attrition_df[X_columns]


# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train['OverTime'].value_counts()
```

```python
#Convert your X data to numeric data types however you see fit
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(dtype="int", sparse_output=False, drop='first', handle_unknown='ignore')
ohe.fit(np.array(X_train['OverTime']).reshape(-1,1))
X_train["Overtime_encoded"] = ohe.transform(np.array(X_train['OverTime']).reshape(-1,1))
X_test["Overtime_encoded"] = ohe.transform(np.array(X_test['OverTime']).reshape(-1,1))
X_train = X_train.drop(columns=['OverTime'])
X_test = X_test.drop(columns=['OverTime'])
```

```python
X_train.head()
```

```python
X_train.dtypes
```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
scaler.fit(X_train, X_test)


# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
y_train['Department'].value_counts()
```

```python
# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(dtype="int", sparse_output=False, drop=None, handle_unknown='ignore')


# Fit the encoder to the training data
ohe.fit(np.array(y_train['Department']).reshape(-1,1))


# Create two new variables by applying the encoder
# to the training and testing data
y_train_depart_encoded = ohe.transform(np.array(y_train['Department']).reshape(-1,1))
y_test_depart_encoded = ohe.transform(np.array(y_test['Department']).reshape(-1,1))
y_train_depart_encoded[:5]
y_test_depart_encoded[:5]
```

```python
# Create a OneHotEncoder for the Attrition column
ohe = OneHotEncoder(dtype="int", sparse_output=False, drop=None, handle_unknown='ignore')



# Fit the encoder to the training data
ohe.fit(np.array(y_train['Attrition']).reshape(-1,1))


# Create two new variables by applying the encoder
# to the training and testing data
y_train_attri_encoded = ohe.transform(np.array(y_train['Attrition']).reshape(-1,1))
y_test_attri_encoded = ohe.transform(np.array(y_test['Attrition']).reshape(-1,1))
print(y_train_attri_encoded[:5])
print(y_test_attri_encoded[:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(X_train.shape[1]), name='input_layer')


# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu', name='shared_layer1')(input_layer)
shared_layer2 = layers.Dense(units=32, activation='relu', name='shared_layer2')(shared_layer1)
```

```python
#Check the shape of 'Department'
y_train_depart_encoded.shape[1]
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
depart_branch = layers.Dense(units=32, activation='relu', name='depart_branch')(shared_layer2)

# Create the hidden layer
depart_hidden = layers.Dense(units=16, activation='relu', name='depart_hidden')(depart_branch)

# Create the output layer
depart_output = layers.Dense(units=3, activation='sigmoid', name='depart_output')(depart_hidden)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attri_branch = layers.Dense(units=32, activation='relu', name='attri_branch')(shared_layer2)
# Create the hidden layer
attri_hidden = layers.Dense(units=16, activation='relu', name='attri_hidden')(attri_branch)

# Create the output layer
attri_output = layers.Dense(units=2, activation='softmax', name='attri_output')(attri_hidden)

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[depart_output, attri_output])


# Compile the model
model.compile(optimizer='adam',
              loss={
                  'depart_output': 'categorical_crossentropy',
                  'attri_output': 'binary_crossentropy'
              },
              metrics={
                  'depart_output': 'accuracy',
                  'attri_output': 'accuracy'
              })


# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled, {'depart_output' : y_train_depart_encoded, 'attri_output' : y_train_attri_encoded}, epochs=10, batch_size=32)


```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled, {'depart_output' : y_test_depart_encoded, 'attri_output' : y_test_attri_encoded})
```

```python
results
```

```python
# Print the accuracy for both department and attrition
print(f'Department Accuracy: {results[3]}')
print(f'Attrition Accuracy: {results[4]}')
```
